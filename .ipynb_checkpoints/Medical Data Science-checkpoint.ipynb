```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from keras.datasets import cifar10
import pandas as pd
import cv2
from skimage import color
from skimage.io import imread_collection
```

## RGB image compression

```python
RGB_img = plt.imread("Flower.jpg")
plt.figure(figsize=(6,7))
plt.imshow(RGB_img)
RGB_img.shape
```

```python
b,g,r = cv2.split(RGB_img)
plt.subplot(1,3,1)
plt.imshow(b)
plt.subplot(1,3,2)
plt.imshow(g)
plt.subplot(1,3,3)
plt.imshow(r)

r_scaled, g_scaled, b_scaled= r/255, g/255, b/255
```

```python
resolution = 0.999
r_pca = PCA(n_components = resolution)
r_pca_projected = r_pca.fit_transform(r_scaled)
g_pca = PCA(n_components = resolution)
g_pca_projected = g_pca.fit_transform(g_scaled)
b_pca = PCA(n_components = resolution)
b_pca_projected = b_pca.fit_transform(b_scaled)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.scatter(range(1,len(r_pca.components_)+1), r_pca.explained_variance_ratio_*100, s = 5)
plt.title("Scree Plot (RED)")
plt.ylabel("% Explained Variance")
plt.xlabel("NO. Components")

plt.subplot(1,3,2)
plt.scatter(range(1,len(g_pca.components_)+1), g_pca.explained_variance_ratio_*100, s = 5)
plt.title("Scree Plot (GREEN)")
plt.ylabel("% Explained Variance")
plt.xlabel("NO. Components")

plt.subplot(1,3,3)
plt.scatter(range(1,len(b_pca.components_)+1), b_pca.explained_variance_ratio_*100, s = 5)
plt.title("Scree Plot (BLUE)")
plt.ylabel("% Explained Variance")
plt.xlabel("NO. Components")
```

```python
pca_r_org = r_pca.inverse_transform(r_pca_projected)
pca_g_org = g_pca.inverse_transform(g_pca_projected)
pca_b_org = b_pca.inverse_transform(b_pca_projected)
img_compressed = cv2.merge((pca_b_org, pca_g_org, pca_r_org))
plt.figure(figsize=(6,7))
plt.imshow(img_compressed)

len(r_pca.components_)
```

# Lung Cancer Scans 

## Preprocessing

```python
train_cancer = imread_collection("archive/lung_cancer_MRI_dataset/train/cancer/*.png")[0:50]
train_no_cancer = imread_collection("archive/lung_cancer_MRI_dataset/train/no_cancer/*.png")[0:50]
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(train_cancer[i])
```

```python
def processImages(images):
    formatted = []
    n = len(images)
    for img in images:
        resized = cv2.resize(img,dsize=(512,512))[:,:,:3] #remove alpha from RGBA
        formatted.append(color.rgb2gray(resized)) #convert to greyscale
    formatted_flattended = np.array(formatted).reshape(n,-1)
    # p = len(formatted_flattended[0])
    # formatted_df = pd.DataFrame(formatted_flattended, columns = ["Pixel{}".format(i) for i in range(1,p+1)])
    # formatted_df['Label'] = 1 if hasCancer else 0
    return formatted_flattended
    
X_cancer = processImages(train_cancer)
Y_cancer = [1 for _ in range(len(X_cancer))]
X_no_cancer = processImages(train_no_cancer)
Y_no_cancer = [0 for _ in range(len(X_no_cancer))]
```

```python
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(X_cancer[i].reshape(512,-1), cmap='grey')
```

```python
X = np.concatenate([X_cancer,X_no_cancer], axis = 0)
Y = np.array(Y_cancer + Y_no_cancer)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
p = len(X_train[0])
```

```python
X_test.shape
```

```python
pca_train = PCA(n_components=50)
# pca_test = PCA(n_components=20)
X_train_projected = pca_train.fit_transform(X_train)
X_train_projected_df = pd.DataFrame(X_train_projected, columns = ["Pixel{}".format(i) for i in range(1,len(X_train_projected[0])+1)])
# X_test_projected = pca_train.fit_transform(X_test)
# X_test_projected_df = pd.DataFrame(X_test_projected, columns = ["Pixel{}".format(i) for i in range(1,len(X_test_projected[0])+1)])
```

```python
inverse = pca_train.inverse_transform(X_train_projected)
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print("Size of training data: {}".format(x_train.shape))
print("Size of test data: {}".format(x_test.shape))
```

```python
labels = np.unique(np.concatenate([y_train,y_test]))
```

```python
label_dict = {
 0: 'airplane',
 1: 'automobile',
 2: 'bird',
 3: 'cat',
 4: 'deer',
 5: 'dog',
 6: 'frog',
 7: 'horse',
 8: 'ship',
 9: 'truck',
}
y_train_mapped = [label_dict[x[0]] for x in y_train]
y_test_mapped = [label_dict[x[0]] for x in y_test]
```

```python
plt.subplot(1,2,1)
plt.imshow(x_train[0])
plt.title("Label: {}".format(y_train_mapped[0]))

plt.subplot(1,2,2)
plt.imshow(x_test[0])
plt.title("Label: {}".format(y_test_mapped[0]))
```

```python
np.min(x_train), np.max(x_train)
```

```python
x_train_normalised = x_train/255
x_train_flat = x_train_normalised.reshape(-1,32*32*3)
columns = ['Pixel' + str(i) for i in range(len(x_train_flat[0]))]
df_cifar = pd.DataFrame(x_train_flat, columns = columns)
df_cifar['label'] = y_train
```

```python
pca_cifar = PCA(n_components = 30)
pca_cifar.fit_transform(df_cifar.iloc[:,:-1])
plt.figure(figsize=(6,5))
```

```python
plt.figure(figsize=(6,7))
plt.scatter(range(1,31),pca_cifar.explained_variance_ratio_)
plt.xlabel("NO. of components")
plt.ylabel("% of Variance")
plt.title("Scree Plot")
```

```python
pca_scree_selected = PCA(n_components = 0.8)
projected_x_train = pca_scree_selected.fit_transform(df_cifar.iloc[:,:-1])
```

```python
n_components = pca_scree_selected.n_components_
projected_df_train = pd.DataFrame(projected_x_train, columns = ["PC{}".format(i) for i in range(1, n_components+1)])
projected_df_train['label'] = y_train
projected_df_train
```

```python
plt.scatter(projected_df_train.iloc[:,0],projected_df_train.iloc[:,1], c = projected_df_train.loc[:,'label'], label = projected_df_train.loc[:,'label'])
plt.legend()
```
